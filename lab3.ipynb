```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Iris.csv")
```

```python
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df, ) 
plt.legend() 

plt.show()
```

```python
sns.pairplot(df.drop(['Id'], axis = 1), hue='Species', height=2)
```

A linear relation as per Species is observed between Petal Length and Petal Width.
No relation between petal length and sepal width.

```python
plt.boxplot(df['SepalLengthCm'])
plt.show()
```

The box plot indicate the quartiles,median, max, min values
Q1=around 5.1  Q3=around 6.4  max=around 7.8  min=around 4.3
Also dataset has no outliers present.

```python
# Select only numerical columns
numerical_df=df.select_dtypes(include=['float64','int64'])
# Compute the correlation matrix
correlation_matrix=numerical_df.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', fmt='.2f', square=True, 
            cbar_kws={"shrink": .8}, linewidths=.5)
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```

Dark shades shows strong bond/relation between two features. Eg Petal Length and Petal Width. 
Light shades show weaker relation. Eg Sepal Width and Sepal Length.

```python
covariance_matrix=df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].cov()
```

```python
print(covariance_matrix)
```

High positive and negative Covariance values indicate direct changes in values between those features. As petal length increases, petal width tends to increase as well.
Positive values show direct relation and negative values show inverse relation between features.
